```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
import matplotlib.lines as mlines
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
def create_map_layout(data, title="Map Title", legend=True, north_arrow=True, scale_bar=True, metadata="Map source: Unknown", border=True):
   
    
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Plot the data (map frame)
    data.plot(ax=ax, color='lightblue', edgecolor='black')
    
    # Set title
    ax.set_title(title, fontsize=16, loc='center')
    
    # Add legend if specified
    if legend:
        handles, labels = ax.get_legend_handles_labels()
        ax.legend(handles, labels, loc='upper left', fontsize=10)

    # Add North Arrow if specified
    if north_arrow:
        ax.annotate('N', xy=(0.9, 0.9), xycoords='axes fraction', fontsize=12, ha='center', va='center')

    # Add scale bar if specified
    if scale_bar:
        scalebar = ScaleBar(1, location='lower right')  # 1 is the scale in meters
        ax.add_artist(scalebar)
    
    # Add metadata (citation) if specified
    ax.text(0.5, -0.05, metadata, ha='center', va='top', transform=ax.transAxes, fontsize=10, color='gray')
    
    # Add border (neatline) if specified
    if border:
        ax.spines['top'].set_visible(True)
        ax.spines['right'].set_visible(True)
        ax.spines['bottom'].set_visible(True)
        ax.spines['left'].set_visible(True)

    # Adjust layout to ensure everything fits
    plt.tight_layout()

    # Show the map
    plt.show()
```

```python
# Example usage:
# Load a sample dataset (replace with your own GeoDataFrame)
```

```python
file_path = r"C:\Users\taabuosi\Downloads\USA (1)\places.shp"
gdf = gpd.read_file(file_path)

# Create a map with desired layout
create_map_layout(gdf, title="USA Places Map", legend=True, north_arrow=True, scale_bar=True, metadata="Source: User's data", border=True)
```
